```python
import pandas as pd
df1=pd.read_csv('/content/sample_data/dataset-1.csv')
df1
```

```python
def generate_car_matrix(data):

    car_matrix = data.pivot(index='id_1', columns='id_2', values='car').fillna(0)


    for index in car_matrix.index:
        car_matrix.at[index, index] = 0

    return car_matrix

```

```python
result_matrix = generate_car_matrix(df1)
print(result_matrix)
```

```python
def get_type_count(data):

    data['car_type'] = pd.cut(data['car'], bins=[float('-inf'), 15, 25, float('inf')],
                              labels=['low', 'medium', 'high'], right=False)


    type_counts = data['car_type'].value_counts().to_dict()


    sorted_type_counts = dict(sorted(type_counts.items()))

    return sorted_type_counts
```

```python
result_dict = get_type_count(df1)
print(result_dict)
```

```python
def get_bus_indexes(df):


  mean_bus = df['bus'].mean()
  filtered_df = df[df['bus'] > 2 * mean_bus]
  return sorted(filtered_df.index.to_list())

```

```python

bus_indexes = get_bus_indexes(df1)
print(bus_indexes)
```

```python
def filter_routes(data):

    filtered_routes = data.groupby('route')['truck'].mean()


    selected_routes = filtered_routes[filtered_routes > 7].index.tolist()


    sorted_routes = sorted(selected_routes)

    return sorted_routes



```

```python
result_routes = filter_routes(df1)
print(result_routes)
```

```python
def multiply_matrix(input_matrix):

    modified_matrix = input_matrix.copy()


    modified_matrix = modified_matrix.applymap(lambda x: x * 0.75 if pd.notna(x) and x > 20 else x * 1.25)


    modified_matrix = modified_matrix.round(1)

    return modified_matrix
```

```python

result_matrix = multiply_matrix(df1)
print(result_matrix)
```

```python
def check_time_completeness(df):

    df['start_timestamp'] = pd.to_datetime(df['startDay'] + ' ' + df['startTime'], errors='coerce')
    df['end_timestamp'] = pd.to_datetime(df['endDay'] + ' ' + df['endTime'], errors='coerce')


    df['start_day_of_week'] = df['start_timestamp'].dt.day_name()
    df['end_day_of_week'] = df['end_timestamp'].dt.day_name()
    df['start_hour'] = df['start_timestamp'].dt.hour
    df['end_hour'] = df['end_timestamp'].dt.hour


    correct_start_timestamps = (
        (df['start_day_of_week'].isin(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])) &
        (df['start_hour'].between(0, 23))
    )
    correct_end_timestamps = (
        (df['end_day_of_week'].isin(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])) &
        (df['end_hour'].between(0, 23))
    )


    correct_timestamps = correct_start_timestamps & correct_end_timestamps


    completeness_series = df.groupby(['id', 'id_2'])['start_timestamp'].transform(lambda x: all(correct_timestamps))

    return completeness_series
```

```python
result_series = check_time_completeness(df2)
print(result_series)
```
